# Dataset Description

This dataset contains global health indicators and demographic data to study factors influencing life expectancy across countries from 2000 to 2015. The factors include comprehensive information such as infant and adult mortality rates, economic factors, disease prevalence, and healthcare indicators.

The dataset has 2938 rows and 22 columns. The columns include the following fields below: 
- Country
- Year
- Status (developing/developed)
- Life expectancy
- adult mortality Rates of both sexes (probability of dying between 15 and 60 years per  1000 population)
- Infant Deaths (number of infant deaths per 1000 population)
- Alcohol, recorded per capita (15+) consumption (in litres of pure alcohol)
- Percentage expenditure on health of Gross Domestic Product per capita (%)
- Hepatitis B (HepB) immunization coverage among 1-year-olds (%)
- Measles - number of reported cases per 1000 population
- BMI - Average Body Mass Index of entire population
- Under-five-deaths - Number of under-five deaths per 1000 population
- Polio (Pol3) immunization coverage among 1-year-olds (%)
- General government expenditure on health as a percentage of total government expenditure (%)
- Diphtheria tetanus toxoid and pertussis (DTP3) immunization coverage among 1-year-olds (%)
- Deaths per 1 000 live births HIV/AIDS (0-4 years)
- Gross Domestic Product per capita (in USD)
- Prevalence of thinness among children and adolescents for Age 10 to 19 (%)
- Prevalence of thinness among children for Age 5 to 9(%)
- Human Development Index in terms of income composition of resources (index ranging from 0 to 1)
- Number of years of Schooling (years)

It has a few data types such as categorical (country, status), integers (year, infant mortality, adult mortality), and floats (BMI, health expenditure). 

This dataset can be accessed via the Kaggle official website. The link to it is https://www.kaggle.com/datasets/kumarajarshi/life-expectancy-who/data.

# Ensuring the Dataset is Within Our Research scope

Since majority of the countries in the dataset contains complete data from 2000 to 2014, we will be using that range to assess the indicators impacting life expectancy. However, some countries do not have data in all of the years in that range hence, we will be taking them out.

```python
# filter rows for the years 2000 to 2014
filteredLifeExpectancy = lifeExpectancyData[(lifeExpectancyData['Year'] >= 2000) & (lifeExpectancyData['Year'] <= 2014)]

# groups data by country and counts the unique years for each country
yearCounts = filteredLifeExpectancy.groupby('Country')['Year'].nunique()

# find countries that do not have data for all years from 2000 - 2014
incompleteCountries = yearCounts[yearCounts < 15].index.tolist()

print(incompleteCountries)
```

```python
# remove rows where the country is in the incompleteCountries list
filteredLifeExpectancy = lifeExpectancyData[~lifeExpectancyData['Country'].isin(incompleteCountries)]

# verify the updated dataset
print(filteredLifeExpectancy.shape)
```

There is still data from 2015 in the dataset however, most of them have null values. Hence, we will remove them as well.

```python
# removes the rows of 2015 data
filteredLifeExpectancy = filteredLifeExpectancy[filteredLifeExpectancy['Year'] != 2015]

print(filteredLifeExpectancy['Year'].max())
print(filteredLifeExpectancy['Year'].min())
```

Some of the countries also do not have a GDP data entry. For the purposes of our project, GDP is essential and hence, we will be excluding the countries that have null GDP values.

```python
# find countries without GDP data
countriesWithoutGDP = filteredLifeExpectancy[filteredLifeExpectancy['GDP'].isnull()]['Country'].unique()

# display the list of countries
print("Countries without GDP data:", countriesWithoutGDP)
```

```python
# remove rows for countries without GDP data
filteredLifeExpectancy = filteredLifeExpectancy[~filteredLifeExpectancy['Country'].isin(countriesWithoutGDP)]

# verify the updated dataset
print(filteredLifeExpectancy.shape)
filteredLifeExpectancy.count()
```

Since this dataset provides a vast number of features that may have impacted life expectancy, we will only focus on a few for our project, such as the country's status, adult and infant mortality rate, immunisation coverage (HepB, Pol3, DTP3), general government expenditure on health, GDP per capita, income composition of resources, and number of years of schooling. As such, we will be removing the other features from our dataset.

```python
# removes whitespaces from the column names
filteredLifeExpectancy.columns = filteredLifeExpectancy.columns.str.strip()

# removes the unwanted columns in our dataset
filteredLifeExpectancy = filteredLifeExpectancy.drop(columns=['Alcohol', 'percentage expenditure', 'Measles', 'BMI', 'under-five deaths', 'HIV/AIDS', 'Population', 'thinness  1-19 years', 'thinness 5-9 years'])

# convert column headers to lowercase
filteredLifeExpectancy.columns = filteredLifeExpectancy.columns.str.lower()

# prints all the column headers to check
print(filteredLifeExpectancy.columns.tolist())

# verifying our dataset
filteredLifeExpectancy.sample(5)
```

```python
filteredLifeExpectancy.count()
```

As we can see above, some of the fields have below 2265 entries. This may be due to multiple reasons. For example, some countries may have only introduced Hepatitis B, Polio, or Diphtheria immunisation later. We now have a more robust dataset with the features that we would like to focus on in this project.

# Statistical Analysis

We will now analyse the key features we have chosen in evaluating the life expectancy of a country. We will be analysing them by finding the statistics below:

### Measures of Central Tendency
Mean: Average value

Median: Middle value after sorting

Mode: Most frequently occurring value

### Measures of Spread
Range: Difference between the min and max values

Variance: How much the data points deviate from the mean

Standard Deviation (SD): Square root of the variance

Interquartile Range (IQR): Range of the middle 50% of the data (difference between the 75th and 25th percentiles)

### Type of Distribution
Symmetric (Normal Distribution): Data is evenly distributed around the mean (0 skewness)

Skewed: Data is lopsided, either left-skewed (negative skewness) or right-skewed (positive skewness)

Multimodal: Data has multiple peaks


## Life Expectancy
### Statistical Summary
In here, we will find the mean, median, and mode of the life expectancy. We will also find its range, variance, standard deviation, and interquartile range. Lastly, we will find its type of distribution.


```python
from scipy.stats import skew
from scipy.stats import kurtosis

print('Life Expectancy Statistics')

# prints out the mean value
print('Mean:', filteredLifeExpectancy['life expectancy'].mean())

# prints out the median value
print('Median:', filteredLifeExpectancy['life expectancy'].median())

# prints out the mode value
print('Mode:', filteredLifeExpectancy['life expectancy'].mode()[0])

# prints out the range
print('Range:', filteredLifeExpectancy['life expectancy'].max() - filteredLifeExpectancy['life expectancy'].min())

# prints out the variance
print('Variance:', filteredLifeExpectancy['life expectancy'].var())

# prints out the standard deviation
print('Standard Deviation:', filteredLifeExpectancy['life expectancy'].std())

# prints out the interquartile range
print('IQR:', filteredLifeExpectancy['life expectancy'].quantile(0.75) - filteredLifeExpectancy['life expectancy'].quantile(0.25))

# measures the asymmetry of the distribution
print('Skewness:', skew(filteredLifeExpectancy['life expectancy']))

# measures the 'tailedness' of the distribution
print('Kurtosis:', kurtosis(filteredLifeExpectancy['life expectancy']))


# prints out the distribution of the data based on its skewness
if skew(filteredLifeExpectancy['life expectancy']) == 0:
    print('Distribution is symmetric')
elif skew(filteredLifeExpectancy['life expectancy']) < 0:
    print('Distribution is left-skewed')
elif skew(filteredLifeExpectancy['life expectancy']) > 0: 
    print('Distribution is right-skewed')
```

## Country Status
Firstly, we will analyse the distribution of developed and developing countries in our dataset. We do this by plotting a pie chart. This will allow us to better envision the percentage of each status.

```python
import matplotlib.pyplot as plt

labels = ['Developing', 'Developed']
colors = ['red', 'green']

# getting the number of countries in the respective statuses
countryStatus = filteredLifeExpectancy['status'].value_counts()

# plotting of pie chart
plt.pie(countryStatus, labels = labels, colors = colors, startangle=90, autopct = '%1.1f%%')

# add title
plt.axis('off')
plt.title('Percentage of Developed and Developing Countries', fontweight = 'bold')

plt.show()
```

From the piechart above, we can see that 81.5% of the countries are still developing and 18.5% of the countries are considered developed. According to the United Nations, developed countries are industrialised, have high standards of living, and have strong economic growth. In contrast, developing countries are agrarian, have lower standards of living, and have very weak economy with slow or non-existent growth.

Let us find out more about how that can affect life expectancy by making a boxplot comparing the two fields.

```python
import seaborn as sns

data = filteredLifeExpectancy

# define custom colors for each category
customPalette = {'Developing': 'red', 'Developed': 'green'}

# create the box plot using seaborn
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid") 
sns.boxplot(x='status', y='life expectancy', data=data, palette=customPalette)

# calculate Q1, median, and Q3 for each category
categories = data['status'].unique()
for category in categories:
    subset = data[data['status'] == category]['life expectancy']
    q1 = subset.quantile(0.25)
    median = subset.median()
    q3 = subset.quantile(0.75)
    
    # find the x-position of the category
    xpos = list(categories).index(category) + 0.52
    
    # annotate Q1
    plt.text(x=xpos, y=q1, s=f'Q1: {q1:.1f}', ha='center', va='center', color='blue', fontsize=10)
    
    # annotate median
    plt.text(x=xpos, y=median, s=f'Median: {median:.1f}', ha='center', va='center', color='black', fontsize=10)
    
    # annotate Q3
    plt.text(x=xpos, y=q3, s=f'Q3: {q3:.1f}', ha='center', va='center', color='blue', fontsize=10)


# add labels and title
plt.title('Life Expectancy by Country Status', fontsize=16)
plt.xlabel('Status', fontsize=12)
plt.ylabel('Life Expectancy (years)', fontsize=12)

# show the plot
plt.show()
```

As we can see from the boxplot above, developing countries have a lower life expectancy than developed countries. The third quartile (Q3) of the developing countries' life expectancy is even lower than the first quartile (Q1) of the developed countries' life expectancy, at 74.1 and 77 years respectively. 

However, the country's status is still a very broad indicator of life expectancy. In the following sections, we will explore more detailed factors that could have affected life expectancy.

## Adult Mortality and Infant Death

In this section, we will be looking at both adult mortality and infant death in developed and developing countries.

From the scatter plot above, we can see that majority of the developing countries had a higher adult mortality than those of the developed countries'. Similarly, there are also generally more infant deaths in developing countries than developed countries, with some going as high as over 1750. 

All of the developed countries have similar statistics of close to 0 infant deaths and slightly more than 50 adult mortality. Meanwhile, developing countries have more varied statistics- from having 0 infant deaths and adult mortality, to having over 1750 and over 700 infant deaths and adult mortality respectively.

# Building the Machine Learning (ML) Model
In this section, we will be building a regression model to help us predict life expectancy.

## Features (Independent Variable)
#### GDP per capita (USD)
This is a measure of economic wealth, and it is often linked to better healthcare infrastructure and higher life expectancy. From our analysis above, we also saw that countries with higher GDP generally had a higher life expectancy.

#### Infant Deaths and Adult Mortality (per 1000 population)
These are strongly correlated with overall health conditions in a country. Adult mortality in particular also indicates the overall health and prevalence of diseases. From our analyses above, we saw that there is a strong relation between these two factors and life expectancy. The developed countries which have a higher life expectancy had a lower infant death and adult mortality count than developing countries which have lower life expectancy.

#### Schooling (years)
Education improves awareness about health practices and access to better jobs and resources. We found that this may have directly impacted the life expectancy through the income and production aspect.

#### Income Composition of Resources (index of 0 - 1)
This is one aspect of the Human Development Index (HDI). The greater the index, the better the capacity of the people to be able to afford resources that improve overall human well-being. We found that this is directly related to education, and hence may have also impacted life expectancy.

## Labels (Dependent Variable)
#### Life Expectancy (years)
This is the target variable we aim to predict, and from our analyses above, we saw that GDP may have a direct relationship with life expectancy.

## Importance of Features
Economic factors like GDP and income composition influence access to healthcare and quality of life. Having higher income or GDP may mean that people in the country will be better able to afford treatments to treat diseases. Additionally, health related metrics like infant deaths and adult mortality directly affect the population's health. Lastly, education impacts long-term health awareness and practices. The more years of education, the greater the knowledge one will have on health.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# load dataset
data = filteredLifeExpectancy

# features and label
features = ['gdp', 'infant deaths', 'adult mortality', 'schooling', 'income composition of resources'] 
label = 'life expectancy'

# drop rows with missing values
data = data.dropna(subset=features + [label])

# split data into X (features) and y (label)
X = data[features]
y = data[label]

# standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# make predictions
y_pred = model.predict(X_test)
```

```python
# evaluate the model

import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))


print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'RMSE: {rmse:.2f}')
print(f'R²: {r2:.2f}')
```

From the results above, we can conclude that the model fits the data. The $R^2$ measures how much variance in the target variable is explained by the model, with the worst being 0 to the best being 1. Given that the above model has an $R^2$ value of 0.76, it means that our model explains a significant portion of variability and has strong correlation. 

Additionally, it has a root mean squared error (RMSE) of 5.09. This means that on average, our model's predictions are off by around 5.09 years only. Since the range of our life expectancy is 52.7 years, this is acceptable.

# Cross Validation

Validation is essential for assessing the robustness of our predictive model. In this section, we will be using the k-fold cross validation to assess our model.

```python
from sklearn.model_selection import cross_val_score

# perform k-fold cross-validation (e.g., k=10)
k = 10
cv_scores = cross_val_score(model, X_scaled, y, cv=k, scoring='r2')

# calculate average R² and standard deviation
mean_cv_score = cv_scores.mean()
std_cv_score = cv_scores.std()

print(f'Average R² from {k}-fold cross-validation: {mean_cv_score:.2f}')
print(f'Standard deviation of R²: {std_cv_score:.2f}')
```

We used the k-fold cross validation as it reduces the risk of overfitting by training and testing the model on different data subsets. It also provides a better estimate of the model's performance on unseen data.

# Feature Engineering

Despite our model being considered good, there is always room for improvement.

```python
# loading the dataset
data = filteredLifeExpectancy

data = data.dropna(subset=['gdp', 'life expectancy'])  # drop rows with missing values in relevant columns

# plotting the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='gdp', y='life expectancy', color='blue', label='Data Points')

# adding title and labels
plt.title('GDP vs Life Expectancy', fontsize=16)
plt.xlabel('GDP (in USD)', fontsize=12)
plt.ylabel('Life Expectancy (years)', fontsize=12)

# show the plot
plt.show()
```

From the graph above, we can see that GDP does not have a linear relationship with life expectancy. Hence, we can include polynomial features to improve our model's performance.

```python
from sklearn.preprocessing import PolynomialFeatures

# apply polynomial features (degree 2)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X_scaled)

# update train-test split
X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)

```

```python
# create and train the linear regression model
model = LinearRegression()
model.fit(X_train_poly, y_train)

# make predictions
y_pred = model.predict(X_test_poly)
```

```python
# evaluate the model
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'RMSE: {rmse:.2f}')
print(f'R²: {r2:.2f}')
```
